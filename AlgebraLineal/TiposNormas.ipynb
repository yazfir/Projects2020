# Tipos de normas: norma 0, norma 1, norma 2, norma infinito y norma L2 al cuadrado

```
La norma (o distancia) se usa en machine learning para encontrar errores.

- La norma L0: devuelve la cantidad de elementos de nuestro vector distintos de cero.

  L0 = #Vi <> 0

- La norma L1: Devuelve la suma sobre i de los valores absolutos de los elementos del vector.

  L1 = { abs(Vi)   donde { : Sumatoria i

- La norma L2: La magnitud del vector desde su origen. También Distancia Euclidiana.

  L2 = Cuánto mide un vector al origen

  (L2)^2 = Misma distancia que L2 pero sin calcular la raíz cuadrada.
        x.dot(x_t)

- La norma infinito: Devuelve el mayor valor dentro de los valores absolutos de nuestro vector.

  L infinito = max abs(Vi)

```

Captura de pantalla de 2021-06-05 21-14-45.png

````python
vector = np.array([1,2,0,5,6,0])

print(np.linalg.norm(vector, ord=0))  # Norma L0
````

Un ejemplo de uso de la norma L0 en el ejemplo anterior podría ser:
¿Cuantas usuarios hicieron compras? Y dentro del vector es la cantidad de cada compra. Con la L0 podemos obtener la respuesta.

````python
vector = np.array([1,-1,1,-1,1])

print(np.linalg.norm(vector, ord=1)) # Norma L1: Suma de todos los valores ABS
````

````python
vector = np.array([1,1])

print(np.linalg.norm(vector)) # Valor por defecto: L2
print(np.linalg.norm(vector, ord=2))  # Norma L2
````

````python
vector = np.array([1,2,3,4,5,6])
print(vector)
````

````python
print(np.linalg.norm(vector, ord=2)**2) # Elevado al cuadrado. (L2)^2
````

````python
print(vector.T.dot(vector)) # (L2)^2
````

````python
vector = np.array([1,2,3,-100])
print(np.linalg.norm(vector, ord=np.inf)) # L infinito : Devuelve valor max absoluto.
````

# Matrices ortogonales y sus propiedades



```
Una matriz es ortogonal cuando todas sus filas son mutuamente ortonormales y todas sus columnas son mutuamente ortonormales.
Así mismo, una matriz ortogonal es una matriz cuadrada cuya matriz inversa coincide con su matriz transpuesta.

Una matriz ortogonal debe cumplir: que todas sus filas sean ortogonales mutuamente, igual que sus columnas y ademas que sean 
ortonormales (Que sus normas sean igual a 1).

Es la matriz cuyos vectores filas son ortonormales mutuamente (lo mismo con los vectores columnas)
Cumplen con la siguiente identidad:
A_tA = AA_t = I
A_t = inv(A)

Nota:
Cuidado con los errores que podemos propagar al operar con las aproximaciones que hace python
```



````python
np.set_printoptions(precision=2, suppress=True) # Ayuda a que si un número es muy cercano a cero no muestre todos los decimales, sólo cero.

matriz = np.array([[1,0,0],
                   [0,1,0],
                   [0,0,1]
                  ])
print(matriz)
````

````python
# Comprobando si los vectores (columnas) dentro de la matriz son Ortonormales

print(matriz[:,0].dot(matriz[:,1])) # v1 a v2

print(matriz[:,0].dot(matriz[:,2])) # v1 a v3

print(matriz[:,1].dot(matriz[:,2])) # v2 a v3
````

````python
print(np.linalg.norm(matriz[:,0]))
print(np.linalg.norm(matriz[:,1]))
print(np.linalg.norm(matriz[:,2]))
````

````python
# Comprobando si los vectores (columnas) dentro de la matriz son Ortonormales

print(matriz[0,:].dot(matriz[1,:]))
print(matriz[0,:].dot(matriz[2,:]))
print(matriz[1,:].dot(matriz[2,:]))
````

````python
print(np.linalg.norm(matriz[0,:]))
print(np.linalg.norm(matriz[1,:]))
print(np.linalg.norm(matriz[2,:]))
````

<h3>Otro Ejemplo</h3>

````python
A = np.array([[np.cos(100), -np.sin(100)],
             [np.sin(100), np.cos(100)]])
print(A)
````

````python
print(np.linalg.norm(A[0,:]))
print(np.linalg.norm(A[1,:]))
print(np.linalg.norm(A[:,0]))
print(np.linalg.norm(A[:,1]))
````

````python
M = A[0,:].dot(A[1,:])
N = A[:,0].dot(A[:,1])

print(round(M,2))
print(round(N,2))
````

````python
A_t = A.T
print(A_t.dot(A))
print(A.dot(A_t))
````

````python
A_inv = np.linalg.inv(A)
print(A_inv)
````

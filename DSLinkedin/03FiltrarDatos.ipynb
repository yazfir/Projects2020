<h1>Herramientas de Filtrado de Pandas</h1>

```python
import pandas as pd
```

```python
df = pd.read_csv('base_datos_2008.csv', nrows=100000)
```

```python
#Obtener informaci√≥n de una columna concreta

df['ArrDelay'].head()
```

```python
#Si quisieramos tomar filas intermedias de muestra

df[100:110] #De la fila 100 a la 110
```

```python
#Si se quiere alg√∫n valor con alguna comparativa concreta
#Ejem. Primeros 5 Vuelos retrasado m√°s 60 min de la lista

df[df['ArrDelay'] > 60].head()
```

```python
#Vuelos que salieron de Houston

df[df["Origin"] == "HOU"].head()
```

```python
# O tambi√©n m√°s complejo como...
# Con & es una y otra (AND)
# Con | es una o otra (OR)

df[ (df["Origin"] == "HOU") & (df['ArrDelay']>60) ].head()
```

```python
#Uso de varios valores de origen en forma de lista

df[df.Origin.isin(['HOU','PHX'])].head()
```

```python
#Con una funci√≥n de pandas que nos dice si no existen los datos (NaN)

df[pd.isna(df['ArrDelay'])].head()
```

```python
#Cantidad de registros o filas sin datos en ArrDelay

len(df[pd.isna(df['ArrDelay'])])
```

<h1>Transformación de Datos</h1>

```python
import pandas as pd
```

```python
df = pd.read_csv('base_datos_2008.csv', nrows=100000)
```

```python
#Crearemos una columna nueva a nuestra BD

df['HoursDelay'] = round(df['ArrDelay']/60)

df['HoursDelay'].head()
```

```python
df.head()
```

```python
#Si queremos borrar una columna

del(df['HoursDelay'])
```

```python
df.head()
```

```python
#Borrar Varias Columnas a la Vez

df.drop(['Diverted','Cancelled','Year'],axis=1) #En este caso sólo lo ejecuta pero no lo guarda en df, habría que igualarlo a df
```

```python
df.head()
```

```python
df.drop(['Diverted','Cancelled','Year'],axis=1, inplace=True) #En este caso lo ejecuta y guarda en df
```

```python
df.head()
```

```python
#Para eliminar filas, por ejem. la fila 0

df.drop(0)
```

```python
#Si se quiere eliminar varias filas, ejem. en un rango

df.drop(range(500,1000)) #Sólo ejecuta no guarda en df, con inplace=True sí lo haría.
```

```python
#Añadir nuevas filas a la BD

dfIND = df[df.Origin == 'IND']
dfHOU = df[df.Origin == 'HOU']
```

```python
newdf = dfIND.append(dfHOU)
```
